```python
#remove Duplicates rows
df=df.drop_duplicates()
```

```python
#Check Data types
df.dtypes
```

```python
#How much total sales did the company generate?
total_sales = df['Sales'].sum()

print(total_sales)
```

```python
#How much total profit?
total_profit = df['Profit'].sum()

print(total_profit)
```

```python
#Average order value
average_sales = df['Sales'].mean()

print(average_sales)
```

```python
#Exploratory Data Analysis (EDA) & Business Insights
```

```python
#Which region generates the highest revenue?
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

print(region_sales)
```

```python
plt.figure(figsize=(8,5))
region_sales.plot(kind='bar')

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.xticks(rotation=0)

plt.show()
```

```python
#Above the graph,The West region contributes the highest sales, while the South region contributes the least.
```

```python
#Which product category earns the most revenue?
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

category_sales
```

```python
plt.figure(figsize=(7,5))
category_sales.plot(kind='bar')

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(rotation=0)

plt.show()
```

```python
#Technology contributes the highest revenue,suggesting strong customer demand for technology products.
```

```python
#Revenue alone isn't enough. Businesses care about profit.
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

category_profit
```

```python
plt.figure(figsize=(7,5))
category_profit.plot(kind='bar')

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.xticks(rotation=0)

plt.show()
```

```python
#Suppose Furniture has high sales but low profit.
```

```python
#Which products generate the most revenue?
top_products = (
    df.groupby('Sub-Category')['Sales']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)
top_products
```

```python
plt.figure(figsize=(12,6))
top_products.plot(kind='bar')

plt.title("Top 10 Products by Sales")
plt.xlabel("Product Name")
plt.ylabel("Sales")
plt.xticks(rotation=0)

plt.show()
```

```python
#The top-selling products account for a significant portion of revenue. Maintaining inventory and ensuring product availability for these items can help sustain sales.
```

```python
#Are highsales regions also highly profitable?
region_profit = (
    df.groupby('Region')['Profit']
      .sum()
      .sort_values(ascending=False)
)

region_profit
```

```python
plt.figure(figsize=(8,5))

region_profit.plot(kind='bar')

plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.xticks(rotation=0)

plt.show()
```

```python
#Are any numerical variables related to each other?
numeric_df = df[['Sales', 'Profit', 'Quantity', 'Discount']]

correlation = numeric_df.corr()

correlation
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(correlation,
            annot=True,
            cmap='coolwarm')
plt.title("Correlation Matrix")

plt.show()
```

```python
#which products are hurting overall profitability?

subcategory_profit = (
    df.groupby('Sub-Category')['Profit']
    .sum()
    .sort_values()
)
subcategory_profit
```
